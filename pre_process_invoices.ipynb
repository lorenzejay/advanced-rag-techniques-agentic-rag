```python
import os
from docling.document_converter import DocumentConverter

document_converter = DocumentConverter()
```

```python

invoices_dir = "knowledge/invoices"
converted = []
for file in os.listdir(invoices_dir):
    if file.endswith(".png"):
        print(f"Processing {file}...")
        docling_doc = document_converter.convert(os.path.join(invoices_dir, file))

        # Extract structured data from the invoice
        export_dict = docling_doc.document.export_to_dict()
        converted.append(export_dict)
        
        
        
        
        
```

```python
print('converted', converted)
```

```python
for invoice in converted:
    print('invoice', invoice)
```

```python
def create_structured_invoice(invoice_data):
    # Extract basic invoice information
    texts = invoice_data.get('texts', [])
    invoice_number = None
    date_of_issue = None
    seller_info = None
    client_info = None
    seller_tax_id = None
    seller_iban = None
    client_tax_id = None
    
    # Extract line items from the first table
    items = []
    if invoice_data.get('tables'):
        items_table = invoice_data['tables'][0]
        table_data = items_table['data']['table_cells']
        current_item = {}
        
        for cell in table_data:
            if cell.get('column_header'):
                continue
                
            if cell.get('text'):
                col_idx = cell['start_col_offset_idx']
                if col_idx == 0 and cell['text'].isdigit():  # New item
                    if current_item:
                        items.append(current_item)
                    current_item = {'item_number': cell['text']}
                elif col_idx == 1:
                    current_item['description'] = cell['text']
                elif col_idx == 2:
                    current_item['quantity'] = cell['text']
                elif col_idx == 4:
                    current_item['unit_price'] = cell['text']
                elif col_idx == 5:
                    current_item['net_worth'] = cell['text']
                elif col_idx == 6:
                    current_item['vat'] = cell['text']
                elif col_idx == 7:
                    current_item['gross_worth'] = cell['text']
        
        if current_item:
            items.append(current_item)

    # Extract summary from second table
    summary = {}
    if len(invoice_data.get('tables', [])) > 1:
        summary_table = invoice_data['tables'][1]
        summary_cells = summary_table['data']['table_cells']
        for cell in summary_cells:
            if cell.get('text') == 'Total':
                row_idx = cell['start_row_offset_idx']
                for total_cell in summary_cells:
                    if total_cell['start_row_offset_idx'] == row_idx:
                        col_idx = total_cell['start_col_offset_idx']
                        if col_idx == 1:
                            summary['total_net'] = total_cell['text']
                        elif col_idx == 2:
                            summary['total_vat'] = total_cell['text']
                        elif col_idx == 3:
                            summary['total_gross'] = total_cell['text']

    # Extract text fields
    for text in texts:
        content = text.get('text', '')
        if 'Invoice no:' in content:
            invoice_number = content.split(': ')[1]
        elif content.count('/') == 2:  # Date format
            date_of_issue = content
        elif text.get('label') == 'section_header':
            if content == 'Seller:':
                seller_info = texts[texts.index(text) + 1]['text']
            elif content == 'Client:':
                client_info = texts[texts.index(text) + 1]['text']
        elif 'Tax Id:' in content:
            if seller_tax_id is None:
                seller_tax_id = content.split(': ')[1]
            else:
                client_tax_id = content.split(': ')[1]
        elif 'IBAN:' in content:
            seller_iban = content.split(': ')[1]

    # Create structured output
    structured_invoice = {
        "invoice_number": invoice_number,
        "date_of_issue": date_of_issue,
        "seller": {
            "name_and_address": seller_info,
            "tax_id": seller_tax_id,
            "iban": seller_iban
        },
        "client": {
            "name_and_address": client_info,
            "tax_id": client_tax_id
        },
        "items": items,
        "summary": summary
    }
    
    return structured_invoice
  
```

```python
structured_invoices = []
for invoice in converted:
    invoice = create_structured_invoice(invoice)
    print('invoice', invoice)
    structured_invoices.append(invoice)

print('structured_invoices', structured_invoices)
```

```python
import os
import weaviate

from dotenv import load_dotenv
from weaviate.classes.config import Configure

load_dotenv()

weaviate_client = weaviate.connect_to_weaviate_cloud(
            cluster_url=os.getenv("WEAVIATE_URL"),
            auth_credentials=weaviate.auth.AuthApiKey(os.getenv("WEAVIATE_API_KEY")),
            headers={"X-OpenAI-Api-Key": os.getenv("OPENAI_API_KEY")},
)
weaviate_client.collections.create(
    "invoices_test",
    vectorizer_config=Configure.Vectorizer.text2vec_openai(),
)
```

```python
collection = weaviate_client.collections.get("invoices_test")
print('structured_invoices', structured_invoices)
with collection.batch.dynamic() as batch:
    for data_row in structured_invoices:
        print('data_row', data_row)
        batch.add_object(
            properties=data_row,
        )
        if batch.number_errors > 10:
            print("Batch import stopped due to excessive errors.")
            break
```

```python
invoices = weaviate_client.collections.get("invoices_test")

response = invoices.query.near_text(
    query="What was the total gross worth of the invoice nunmber 13194726?",
    limit=2
)

print(response.objects[0].properties) 
```
